# Importing Libraries

---



```python
import pandas as pd
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix as con
from sklearn.metrics import classification_report
```

```python
sample = datasets.load_iris()
```

```python
X = pd.DataFrame(sample.data, columns = sample.feature_names)
```

```python
X.info()
```

```python
X.head(5)
```

```python
X.describe()
```

```python
Y = sample.target
print(Y)
```



---


# Spliting Train and Test Data


---



```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```



---


# Decision Tree Algorithm


---



```python
decision = DecisionTreeClassifier()
decision.fit(X_train,Y_train)
```

```python
Y_predict = decision.predict(X_test)
```

```python

feature=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cls=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (16,10), dpi = 300)

tree.plot_tree(decision, feature_names = feature, class_names = cls, filled = True);
```



---


# Evaluating The Decision Tree Model Using Confusion Matrix And Classification Report


---



```python
con(Y_test, Y_predict)
```

```python
report = classification_report(Y_test,Y_predict, labels=[0,1,2], target_names=['setosa','versicolor','virginica'])
print('Classification report : \n', report)
```
